```python
#creating Node class
class Node:
  def __init__(self, element, next):
    self.element = element
    self.next = next
```

```python
#Stack class
class Stack:
  def __init__(self):
    self.top = None
  #Push operation
  def push(self, elem):
    if self.top == None:
      self.top = Node(elem, None)
    else:
      n = Node(elem, None)
      n.next = self.top
      self.top = n
  #Pop operation
  def pop(self):
    if self.top == None:
      return None
    else:
      n = self.top
      self.top = self.top.next
      n.next = None
      return n.element
  #Peek operation
  def peek(self):
    if self.top == None:
      return None
    else:
      n = self.top.element
      return n
  #isEmpty operation
  def isEmpty(self):
    if self.top == None:
      return True
    else:
      return False
```

```python
def print_stack(st):
  if st.isEmpty():
    return None
  else:
    n = st.pop()
    print(n)
    print_stack(st)
    st.push(n)

st = Stack()
for i in range(1, 10):
  st.push(i)
print_stack(st)
```

```python
def infix(st, arr):
  for i in range(len(arr)):
    if arr[i] == "+" or arr[i] == "-" or arr[i] == "*" or arr[i] == "/":
      if st.isEmpty() != True:
        x = int(st.pop())
        y = int(st.pop())
        if arr[i] == "+":
          st.push(y+x)
        elif arr[i] == "-":
          st.push(y-x)
        elif arr[i] == "*":
          st.push(y*x)
        elif arr[i] == "/":
          st.push(y/x)
    else:
      st.push(arr[i])
  return st.pop()

st = Stack()
arr = ["3", "2", "+", "4", "*", "5", "1", "-", "/"]
infix(st, arr)
```
