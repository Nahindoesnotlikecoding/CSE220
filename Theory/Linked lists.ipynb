## Linked List
* We'll make objects which will hold 2 attributes. These are 1. value 2. location of the next object. These objects are called "Nodes".
* This creates an invisible link among the nodes. That's why, this is called "Linked List".
* The first node reference is called a "Head". it consists of the location of the first node.
* Linked Lists do not allow random access whereas arrays do.

```python
import numpy as np
```

```python
class linkedList:
  def __init__(self, arr):
    self.head = Node(arr[0], None)
    tail = self.head
    for i in range(1, len(arr)):
        x = Node(arr[i], None)
        tail.next = x
        tail = x

class Node:
  def __init__(self, value, next):
    self.value = value
    self.next = next
```

```python
class linkedList:
  def __init__(self, arr):
    self.head = Node(arr[0], None)
    tail = self.head
    for i in range(1, len(arr)):
        x = Node(arr[i], None)
        tail.next = x
        tail = x

class Node:
  def __init__(self, value, next):
    self.value = value
    self.next = next

arr = np.array([10, 20, 30, 40, 50])
obj = linkedList(arr)
print(obj.head.value)
```

```python
# Print value of all the nodes
def printNode(head):
  n = head
  while n != None:
    print(n.value, end=" ")
    n = n.next

arr = np.array([10, 20, 30, 40, 50])
obj = linkedList(arr)
printNode(obj.head)
```

```python
# Count how many nodes are there
def countNode(head):
  n = head
  count = 0
  while n != None:
    count += 1
    n = n.next
  return count

arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])
obj = linkedList(arr)
print(countNode(obj.head))
```

```python
# Get the location of a node with respect to a given index
def getNode(head, index):
  count = 0
  temp = head
  while temp != None:
    if count == index:
      return temp
    else:
      temp = temp.next
      count += 1
  return f"Not a valid index"


arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])
obj = linkedList(arr)
print(getNode(obj.head, 0))
print(getNode(obj.head, 1))
print(getNode(obj.head, 2))
print(getNode(obj.head, -1))
```

```python
#get the element of a node with respect to the given index
def getElement(head, index):
  temp = head
  count = 0
  while temp != None:
    if count == index:
      return temp.value
    else:
      temp = temp.next
      count += 1
  return f"Not a valid index"

arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])
obj = linkedList(arr)
print(getElement(obj.head, 0))
print(getElement(obj.head, 5))
print(getElement(obj.head, 9))
print(getElement(obj.head, -1))
```

```python
#Update value of a node at a given index
def update(head, index, element):
  temp = head
  count = 0
  while temp != None:
    if count == index:
      temp.value = element
      break
    else:
      temp = temp.next
      count += 1

arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])
obj = linkedList(arr)
print(getElement(obj.head, 2))
update(obj.head, 2, 100)
print(getElement(obj.head, 2))
```

```python
#getting index of a specific element
def search(head, element):
  temp = head
  count = 0
  while temp != None:
    if temp.value == element:
      return count
    else:
      temp = temp.next
      count += 1
  return f"The element is not found"

arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])
obj = linkedList(arr)
print(search(obj.head, 50))
```

## Insert
Insertion happens in 4 different cases
* insert at the beginning
* insert at the end
* insert at the middle
* insert out of bound

```python
def insert(head, index, element):
  total_nodes = countNode(head)

  if index == 0:                        #Inserting at the beginning
    n = Node(element, head)
    head = n

  elif index>0 and index<total_nodes:        #Inserting at the middle
    previous_node = getNode(head, index-1)
    next_node = getNode(head, index)
    n = Node(element, next_node)
    previous_node.next = n

  elif index==total_nodes:                     #Inserting at the end
    n = Node(element, None)
    previous_node = getNode(head, index-1)
    previous_node.next = n

  else:
    print(f"Index out of bound")

  return head

arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])
obj = linkedList(arr)
head = insert(obj.head, 8, 100)

print(head)
printNode(head)
```

```python
#Removing an element also happens like insertion
def remove(head, index):
  total_nodes = countNode(head)

  if index == 0:
    head = head.next
  elif index>0 and index<total_nodes:
    previous_node = getNode(head, index-1)
    next_node = getNode(head, index)
    previous_node.next = next_node.next
  else:
    print("Invalid index")
  return head

arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])
obj = linkedList(arr)
printNode(obj.head)
head = remove(obj.head, 3)
print()
printNode(head)
```

```python
#Removing an element also happens like insertion
#This is my way to understand simply
def remove(head, index):
  total_nodes = countNode(head)

  if index == 0:
    head = head.next
  elif index>0 and index<total_nodes-1:
    previous_node = getNode(head, index-1)
    next_node = getNode(head, index+1)
    previous_node.next = next_node
  elif index==total_nodes-1:
    previous_node = getNode(head, index-1)
    previous_node.next = None
  else:
    print("Invalid index")
  return head

arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])
obj = linkedList(arr)
printNode(obj.head)
head = remove(obj.head, 7)
print()
printNode(head)
```

```python
#Right Rotate a linked list
def rightRotate(head):
  total_nodes = countNode(head)
  last_node = getNode(head, total_nodes-1)
  second_last_node = getNode(head, total_nodes-2)
  last_node.next = head
  second_last_node.next = None
  head = last_node
  return head

arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])
obj = linkedList(arr)
head = rightRotate(obj.head)
printNode(head)
```

```python
#Left rotate a linked list
def leftRotate(head):
  total_nodes = countNode(head)
  last_node = getNode(head, total_nodes-1)
  second_node = getNode(head, 1)
  last_node.next = head
  head.next = None
  head = second_node
  return head

arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])
obj = linkedList(arr)
head = leftRotate(obj.head)
printNode(head)
```

```python
#reverse a linked list (out of place)
def reverse(head):
  new_head = Node(head.value, None)
  temp = head.next
  while temp != None:
    n = Node(temp.value, new_head)
    new_head = n
    temp = temp.next
  return new_head

arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])
obj = linkedList(arr)
head = reverse(obj.head)
printNode(head)
```

```python
#reverse a linked list (in place)
def reverse(head):
  temp = head.next
  head.next = None
  while temp != None:
    temp2 = temp.next
    temp.next = head
    head = temp
    temp = temp2
  return head

arr = np.array([1, 2, 3, 4, 5])
obj = linkedList(arr)
new_head = reverse(obj.head)
printNode(new_head)
```

## Dummy headed doubly circular linked list

```python
#Dummy headed circular doubly linked list creation
class doublyNode:
  def __init__(self, value, next, prev):
    self.value = value
    self.next = next
    self.prev = prev

def doublyLinkedList(arr):
  dh = doublyNode(None, None, None)
  dh.next = dh
  dh.prev = dh
  tail = dh
  for i in range(len(arr)):
    n = doublyNode(arr[i], dh, tail)
    tail.next = n
    tail = tail.next
    dh.prev = tail
  return dh

arr = np.array([1, 2, 3, 4, 5])
obj = doublyLinkedList(arr)
```

```python
def printNode(dh):
  temp = dh.next
  while temp != dh:
    print(temp.value, end=" ")
    temp = temp.next

arr = np.array([1, 2, 3, 4, 5])
dh = doublyLinkedList(arr)
printNode(dh)
```

```python
def countNode(dh):
  temp = dh.next
  count = 0
  while temp != dh:
    count += 1
    temp = temp.next
  return count

arr = np.array([1, 2, 3, 4, 5])
dh = doublyLinkedList(arr)
print(countNode(dh))
```

```python
def getNode(dh, index):
  temp = dh.next
  count = 0
  while temp != dh:
    if count == index:
      return temp
    else:
      count += 1
      temp = temp.next
  return f"Not a valid index"


arr = np.array([1, 2, 3, 4, 5])
dh = doublyLinkedList(arr)
print(getNode(dh, 0))
x = getNode(dh, 1)
print(x.value)
```

```python
def insertion(dh, value, index):
  next_node = getNode(dh, index)
  prev_node = next_node.prev
  n = doublyNode(value, next_node, prev_node)
  prev_node.next = n
  next_node.prev = n

arr = np.array([1, 2, 3, 4, 5])
dh = doublyLinkedList(arr)
insertion(dh, 99, 1)
printNode(dh)
```

```python
def remove(dh, index):
  removed_node = getNode(dh, index)
  next_node = removed_node.next
  prev_node = removed_node.prev
  prev_node.next = next_node
  next_node.prev = prev_node

arr = np.array([1, 2, 3, 4, 5])
dh = doublyLinkedList(arr)
remove(dh, 2)
printNode(dh)
```

```python
def reverse(dh):
  temp = dh.next
  dh.next = dh.prev
  dh.prev = temp
  while temp != dh:
    temp2 = temp.next
    temp.next = temp.prev
    temp.prev = temp2
    temp = temp2
  return dh

arr = np.array([1, 2, 3, 4, 5, 6, 7])
dh = doublyLinkedList(arr)
dh = reverse(dh)
printNode(dh)
```

```python
def insert(dh, elem, index):
  total_nodes = countNode(dh)
  if index == total_nodes:
    prev_node = getNode(dh, index-1)
    n = doublyNode(elem, None, prev_node)
    prev_node.next = n
  else:
    next_node = getNode(dh, index)
    prev_node = next_node.prev
    n = doublyNode(elem, next_node, prev_node)
    prev_node.next = n
    next_node.prev = n


arr = np.array([1, 2, 3, 4, 5])
dh = doublyLinkedList(arr)
insert(dh, 100, 5)
printNode(dh)
```

```python
def remove(dh, index):
  removed_node = getNode(dh, index)
  total_nodes = countNode(dh)
  if index < total_nodes-1:
     prev_node = removed_node.prev
     next_node = removed_node.next
     prev_node.next = next_node
     next_node.prev = prev_node
  elif index == total_nodes-1:
    prev_node = removed_node.prev
    prev_node.next = None

arr = np.array([1, 2, 3, 4, 5])
dh = doublyLinkedList(arr)
remove(dh, 4)
printNode(dh)
```

## Non Dummy Headed Doubly Circular Linked List

```python
class Node:
  def __init__(self, elem, next, prev):
    self.elem = elem
    self.next = next
    self.prev = prev

def doublyLinkedList(arr):
  head = Node(arr[0], None, None)
  head.next = head
  head.prev = head
  tail = head
  for i in range(1, len(arr)):
    n = Node(arr[i], head, tail)
    tail.next = n
    tail = tail.next
    head.prev = tail
  return head

arr = np.array([1, 2, 3, 4, 5])
head = doublyLinkedList(arr)
print(head)
```

```python
def printNode(head):
  temp = head.next
  print(head.elem, end=" ")
  while temp != head:
    print(temp.elem, end=" ")
    temp = temp.next

arr = np.array([1, 2, 3, 4, 5])
head = doublyLinkedList(arr)
printNode(head)
```
