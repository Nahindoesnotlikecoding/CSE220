```python
import numpy as np
```

```python
class BTNode:
  def __init__(self, elem, left=None, right=None):
    self.elem = elem
    self.left = left
    self.right = right
```

```python
def inorder(root):
  if root == None:
    return None
  inorder(root.left)
  print(root.elem, end=" ")
  inorder(root.right)
```

```python
def createBST(arr, l, r):
  if l > r:
    return None
  mid = (l+r)//2
  root = BTNode(arr[mid])

  root.left = createBST(arr, l, mid-1)
  root.right = createBST(arr, mid+1, r)
  return root


arr = np.array([1, 2, 3, 4, 5, 6, 7])
root = createBST(arr, 0, len(arr)-1)
inorder(root)
```

```python
def height(root):
  if root == None:
    return -1

  return 1 + max(height(root.left), height(root.right))


root = BTNode(10)
root.left = BTNode(4)
root.right = BTNode(6)
root.left.left = BTNode(4)
root.left.right = BTNode(3)
root.right.left = BTNode(1)
root.right.right = BTNode(3)
root.right.right.left = BTNode(2)
root.right.right.right = BTNode(1)
root.right.right.right.right = BTNode(6)

x = height(root)
print(x)
```

```python
def insert(root, value):
  if root == None:
    return None
  elif root.elem > value:
    if root.left == None:
      root.left = BTNode(value)
    else:
      insert(root.left, value)
  elif root.elem < value:
    if root.right == None:
      root.right = BTNode(value)
    else:
      insert(root.right, value)
  else:
    print("Duplicate!!")

root = BTNode(70)
root.left = BTNode(50)
root.right = BTNode(90)
root.left.left = BTNode(40)
root.left.right = BTNode(60)
root.left.left.left = BTNode(20)
root.right.left = BTNode(80)
root.right.right = BTNode(95)
root.right.left.left = BTNode(75)
root.right.left.right = BTNode(85)
root.right.right.right = BTNode(99)

inorder(root)
print()

insert(root, 10)
inorder(root)
print()

insert(root, 30)
inorder(root)
print()

insert(root, 25)
inorder(root)
print()
```

```python
def predecessor(root, value):
  if root == None:
    return None
  elif root.elem >= value:
    return predecessor(root.left, value)
  else:
    ans = predecessor(root.right, value)
    if ans != None:
      return ans
    else:
      return root.elem

root = BTNode(70)
root.left = BTNode(50)
root.right = BTNode(90)
root.left.left = BTNode(40)
root.left.right = BTNode(60)
root.left.left.left = BTNode(20)
root.right.left = BTNode(80)
root.right.right = BTNode(95)
root.right.left.left = BTNode(75)
root.right.left.right = BTNode(85)
root.right.right.right = BTNode(99)
inorder(root)
print()

x = predecessor(root, 80)
print(x)
```

```python
def successor(root, value):
  if root == None:
    return None
  elif root.elem <= value:
    return successor(root.right, value)
  else:
    ans = successor(root.left, value)
    if ans != None:
      return ans
    else:
      return root.elem

root = BTNode(70)
root.left = BTNode(50)
root.right = BTNode(90)
root.left.left = BTNode(40)
root.left.right = BTNode(60)
root.left.left.left = BTNode(20)
root.right.left = BTNode(80)
root.right.right = BTNode(95)
root.right.left.left = BTNode(75)
root.right.left.right = BTNode(85)
root.right.right.right = BTNode(99)
inorder(root)
print()

x = successor(root, 80)
print(x)
```

```python
# delete with the help of predecessor
def delete(root, value):
  if root == None:
    return None
  elif root.elem > value:
    root.left = delete(root.left, value)
    return root
  elif root.elem < value:
    root.right = delete(root.right, value)
    return root
  else:
    if root.left == None and root.right == None:
      return None
    elif root.left == None:
      return root.right
    elif root.right == None:
      return root.left
    predecessor = root.left
    while(predecessor.right != None):
      predecessor = predecessor.right

    root.elem = predecessor.elem
    root.left = delete(root.left, predecessor.elem)
    return root


root = BTNode(70)
root.left = BTNode(50)
root.right = BTNode(90)
root.left.left = BTNode(40)
root.left.right = BTNode(60)
root.left.left.left = BTNode(20)
root.right.left = BTNode(80)
root.right.right = BTNode(95)
root.right.left.left = BTNode(75)
root.right.left.right = BTNode(85)
root.right.right.right = BTNode(99)
inorder(root)
print()

delete(root, 90)
inorder(root)
```

```python
# delete with the help of successor
def delete(root, value):
  if root == None:
    return None
  elif root.elem > value:
    root.left = delete(root.left, value)
    return root
  elif root.elem < value:
    root.right = delete(root.right, value)
    return root
  else:
    if root.left == None and root.right == None:
      return None
    elif root.left == None:
      return root.right
    elif root.right == None:
      return root.left
    successor = root.right
    while(successor.left != None):
      successor = successor.left

    root.elem = successor.elem
    root.right = delete(root.right, successor.elem)
    return root


root = BTNode(70)
root.left = BTNode(50)
root.right = BTNode(90)
root.left.left = BTNode(40)
root.left.right = BTNode(60)
root.left.left.left = BTNode(20)
root.right.left = BTNode(80)
root.right.right = BTNode(95)
root.right.left.left = BTNode(75)
root.right.left.right = BTNode(85)
root.right.right.right = BTNode(99)
inorder(root)
print()

delete(root, 90)
inorder(root)
```

```python
# Count numbers in a range
def numsInrange(root, x, y):
  if root == None:
    return 0
  elif x <= root.elem <= y:
    return 1+numsInrange(root.left, x, y) + numsInrange(root.right, x, y)
  else:
    if root.elem < x:
      return numsInrange(root.right, x, y)
    else:
      return numsInrange(root.left, x, y)

root = BTNode(10)
root.left = BTNode(5)
root.right = BTNode(15)
root.left.left = BTNode(1)
root.left.right = BTNode(7)
root.right.left = BTNode(12)
root.right.right = BTNode(7)
x = numsInrange(root, 10, 15)
print(x)
```

```python
# closest element in BST
def closest(root, x):
  insert(root, x)
  pred = predecessor(root, x)
  succ = successor(root, x)
  if pred == None or succ== None:
    if pred==None:
      return succ
    else:
      return pred
  elif (x-pred) < (succ-x):
    return pred
  else:
    return succ

root = BTNode(10)
root.left = BTNode(5)
root.right = BTNode(15)
root.left.left = BTNode(1)
root.left.right = BTNode(7)
root.right.left = BTNode(12)
root.right.right = BTNode(17)
x = numsInrange(root, 10, 15)
x = closest(root, 18)
print(x)
```

```python
#kth smallest element
def leftestroot(root):
  if root.left == None:
    return root
  x = leftestroot(root.left)
  return x

def kthsmallest(root, k):
  left = leftestroot(root)
  x = left.elem
  for i in range(k-1):
    x = successor(root, x)
  return x

root = BTNode(10)
root.left = BTNode(5)
root.right = BTNode(15)
root.left.left = BTNode(1)
root.left.right = BTNode(7)
root.right.left = BTNode(12)
root.right.right = BTNode(17)
x = kthsmallest(root, 4)
print(x)
```

```python
#kth largest element
def rightestroot(root):
  if root.right == None:
    return root
  x = rightestroot(root.right)
  return x

def kthlargest(root, k):
  right = rightestroot(root)
  x = right.elem
  for i in range(k-1):
    x = predecessor(root, x)
  return x

root = BTNode(10)
root.left = BTNode(5)
root.right = BTNode(15)
root.left.left = BTNode(1)
root.left.right = BTNode(7)
root.right.left = BTNode(12)
root.right.right = BTNode(17)
x = kthlargest(root, 1)
print(x)
```
