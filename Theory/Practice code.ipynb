```python
class Node:
  def __init__(self, element, next):
    self.element = element
    self.next = next

class Stack:
  def __init__(self):
    self.top = None

  def push(self, elem):
    if self.top == None:
      n = Node(elem, None)
      self.top = n
    else:
      n = Node(elem, None)
      n.next = self.top
      self.top = n
  def pop(self):
    if self.top == None:
      return None
    else:
      n = self.top
      self.top = self.top.next
      n.next = None
      return n.element
  def peek(self):
    if self.top == None:
      return None
    else:
      return self.top.element
  def isEmpty(self):
    if self.top == None:
      return True
    else:
      return False

st = Stack()
st.push(4)
st.push(3)
st.push(5)
st.push(1)
st.push(9)
```

```python
def infix(st, arr):
  for i in range(len(arr)):
    if arr[i] == "+" or arr[i] == "-" or arr[i] == "*" or arr[i] == "/":
      if st.isEmpty() != True:
        x = int(st.pop())
        y = int(st.pop())
        if arr[i] == "+":
          st.push(y+x)
        elif arr[i] == "-":
          st.push(y-x)
        elif arr[i] == "*":
          st.push(y*x)
        elif arr[i] == "/":
          st.push(y/x)
    else:
      st.push(arr[i])
  return st.pop()

st = Stack()
arr = ["3", "2", "+", "4", "*", "5", "1", "-", "/"]
infix(st, arr)
```

```python
def insert(head, elem, index):
  total_nodes = countNode(head)
  if index == 0:
    n = Node(elem, head)
    head = n
  elif 0<index<total_nodes:
    previous_node = getNode(head, index-1)
    next_node = getNode(head, index)
    n = Node(elem, next_node)
    previous_node.next = n
  elif index == total_nodes:
    previous_node = getNode(head, index-1)
    n = Node(elem, None)
    previous_node.next = n
  else:
    return f"Index out of bound"
  return head

arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])
head = linkedList(arr)
head = insert(head, 90, 5)
printNode(head)
```

```python
class MinHeap:
  def __init__(self, cap):
    self.arr = np.zeros(cap, dtype="int")
    self.size = 0
    self.cap = cap

  def insert(self, value):
    if self.size < self.cap:
      self.size += 1
      self.arr[self.size-1] = value
      self.swim(self.size-1)
    else:
      return

  def swim(self, index):
    if index>0:
      parent = (index-1)//2
      if self.arr[index] < self.arr[parent]:
        self.arr[index], self.arr[parent] = self.arr[parent], self.arr[index]
        self.swim(parent)
      else:
        return

  def extractMin(self):
    temp = self.arr[0]
    self.arr[0], self.arr[self.size-1] = self.arr[self.size-1], self.arr[0]
    self.size-=1
    self.arr[self.size] = 0
    self.sink(0)
    return temp

  def sink(self, index):
    if index<self.size:
      left = (2*index)+1
      right = (2*index)+2
      if left < self.size:
        if right < self.size:
          small = 0
          if self.arr[left] < self.arr[right]:
            small = left
          else:
            small = right
          if self.arr[small] < self.arr[index]:
            self.arr[small], self.arr[index] = self.arr[index], self.arr[small]
            self.sink(small)
        else:
          if self.arr[left] < self.arr[index]:
            self.arr[left], self.arr[index] = self.arr[index], self.arr[left]
            self.sink(left)

    return

  def heapSort(self, inp_arr):
    sort_heap = MinHeap(len(inp_arr))
    for i in range(len(inp_arr)):
      sort_heap.insert(inp_arr[i])
    for i in range(len(inp_arr)):
      extracted = sort_heap.extractMin()
      inp_arr[i] = extracted

    return inp_arr

  def getArray(self):
    return self.arr
```

```python
#Stack class
class Stack:
  def __init__(self):
    self.top = None
  #Push operation
  def push(self, elem):
    if self.top == None:
      self.top = Node(elem, None)
    else:
      n = Node(elem, None)
      n.next = self.top
      self.top = n
  #Pop operation
  def pop(self):
    if self.top == None:
      return None
    else:
      n = self.top
      self.top = self.top.next
      n.next = None
      return n.elem
  #Peek operation
  def peek(self):
    if self.top == None:
      return None
    else:
      n = self.top.elem
      return n
  #isEmpty operation
  def isEmpty(self):
    if self.top == None:
      return True
    else:
      return False
```

```python
class MinHeap:
  def __init__(self, cap):
    self.__arr = np.zeros(cap, dtype="int")
    self.__size = 0
    self.__cap = cap

  def insert(self, value):
    if self.__size < self.__cap:
      self.__size += 1
      self.__arr[self.__size-1] = value
      self.swim(self.__size-1)
    return

  def swim(self, index):
    if index > 0:
      parent = (index-1)//2
      if self.__arr[parent] > self.__arr[index]:
        self.__arr[parent], self.__arr[index] = self.__arr[index], self.__arr[parent]
        self.swim(parent)
      else:
        return

  def extractMin(self):
    temp = self.__arr[0]
    self.__arr[0], self.__arr[self.__size-1] = self.__arr[self.__size-1], self.__arr[0]
    self.__size -= 1
    self.__arr[self.size] = 0
    self.sink(0)
    return temp

  def sink(self, index):
    if index<self.__size:
      left = (2*index)+1
      right = (2*index)+2
      if left < self.__size:
        if right < self.__size:
          small = 0
          if self.__arr[left] < self.__arr[right]:
            small = left
          else:
            small = right
          if self.__arr[small] < self.__arr[index]:
            self.__arr[small], self.__arr[index] = self.__arr[index], self.__arr[small]
            self.sink(small)
        else:
          if  self.__arr[left] < self.__arr[index]:
            self.__arr[left], self.__arr[index] = self.__arr[index], self.__arr[left]
            self.sink(left)
    return

  def getArray(self):
    return self.__arr

```

```python
class Node:
  def __init__(self, elem, next=None):
    self.elem = elem
    self.next = next

def linkedList(arr):
  head = Node(arr[0])
  temp = head
  for i in range(1, len(arr)):
    n = Node(arr[i])
    temp.next = n
    temp = temp.next

  return head

def countNode(head):
  temp = head
  count = 0
  while temp!=None:
    count += 1
    temp = temp.next

  return count
def getNode(head, index):
  temp = head
  count = 0
  while temp != None:
    if count == index:
      return temp
    count += 1
    temp = temp.next

def printNode(head):
  temp = head
  while temp != None:
    print(temp.elem, end=" --> ")
    temp = temp.next
  print("None")

arr = np.array(["A3", "A9", "A4", "A2", "A7", "A8", "A1"])
head = linkedList(arr)
printNode(head)
```
