```python
#Assignment Part 1
class Patient:
  #write a constructor
  def __init__(self, id, name, age, bloodgroup, next, prev):
    self.id = id
    self.name = name
    self.age = age
    self.bloodgroup = bloodgroup
    self.next = next
    self.prev = prev
```

```python
class WRM:
  def __init__(self):
    #Creating the dummy head
    self.dh = Patient(None,None,None,None,None,None)
    self.dh.next = self.dh
    self.dh.prev = self.dh

  def registerPatient(self,id, name, age, bloodgroup):
    n = Patient(id, name, age, bloodgroup, self.dh, self.dh.prev)
    self.dh.prev.next = n
    self.dh.prev = n

  def servePatient(self):
    if self.dh.next == self.dh:
      return "No patient to serve"
    else:
      temp = self.dh.next
      name = temp.name
      self.dh.next = temp.next
      temp.next.prev = self.dh
      print(f"{name} is served.")

  def showAllPatient(self):
    temp = self.dh.next
    while temp != self.dh:
      print(temp.name, end=" ")
      temp = temp.next

  def canDoctorGoHome(self):
    if self.dh.next == self.dh:
      return True
    else:
      return False

  def cancelAll(self):
    self.dh.next = self.dh
    self.dh.prev = self.dh

  def ReverseTheLine(self):
    temp = self.dh.next
    self.dh.next = self.dh.prev
    self.dh.prev = temp
    while temp != self.dh:
      temp2 = temp.next
      temp.next = temp.prev
      temp.prev = temp2
      temp = temp2
```

```python
#Write a Tester Code in this cell
print("**Welcome to Waiting Room Management System**")
def_print =   """
==Choose an Option==
1. RegisterPatient()
2. ServePatient()
3. CancelAll()
4. CanDoctorGoHome()
5. ShowAllPatient()
6. ReverseTheLine()
7. exit
=============================
"""
patient = WRM()
while True:
  print(def_print)
  x = int(input("Enter your choice: "))
  if x==1:
    print("executing RegisterPatient()... ")
    id = int(input("Enter ID: "))
    name = input("Enter Name: ")
    age = int(input("Enter Age: "))
    bg = input("Enter Blood Group: ")
    patient.registerPatient(id, name, age, bg)
    print("Success registering patient")
  elif x==2:
    print("executing servePatient()...")
    patient.servePatient()
  elif x==3:
    print("executing cancelAll()...")
    patient.cancelAll()
    print("All appointment have been canceled.")
  elif x == 4:
    print("executing canDoctorGoHome()...")
    if patient.canDoctorGoHome():
      print("Yes")
    else:
      print("No")
  elif x==5:
    print("executing showAllPatient()...")
    if patient.dh.next == patient.dh:
      print("No patient.")
    else:
      patient.showAllPatient()
  elif x==6:
    print("executing reverseTheLine()...")
    patient.ReverseTheLine()
  elif x==7:
    break
  else:
    print("No such option")
```

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

```python
def print_stack(st):
  if st.isEmpty():
    return
  p = st.pop()
  print('|',p,end=' ')
  if p<10:
    print(' |')
  else:
    print('|')
  #print('------')
  print_stack(st)
  st.push(p)
```

```python
def diamond_count(Stack, string):
  count = 0
  for i in range(len(string)):
    if string[i] == "<" or string[i] == ">":
      if string[i] == "<":
        Stack.push(string[i])
      else:
        if Stack.isEmpty():
          Stack.push(string[i])
        else:
          n = Stack.peek()
          if n=="<":
            Stack.pop()
            count += 1
  return count


print('Test 01')
stack = Stack()
string = '<..><.<..>> '
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')

print('Test 02')
stack = Stack()
string = '<<<..<......<<<<....>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 1
unittest.output_test(returned_value, 1)
print('-----------------------------------------')


print('Test 03')
stack = Stack()
string = '>>><...<<..>>...>...>>>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')
```

```python
def remove_block(stack, n):
  obj = Stack()
  for i in range(n-1):
    n = stack.pop()
    obj.push(n)
  stack.pop()
  while obj.isEmpty() != True:
    n = obj.pop()
    stack.push(n)
  return stack

print('Test 01')
st = Stack()
st.push(4)
st.push(19)
st.push(23)
st.push(17)
st.push(5)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,2)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()

print('Test 02')
st = Stack()
st.push(73)
st.push(85)
st.push(15)
st.push(41)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,3)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()
```

```python
def conditional_reverse(stack):
  obj = Stack()
  while stack.isEmpty() != True:
    n = stack.pop()
    if obj.isEmpty():
      obj.push(n)
    else:
      peek = obj.peek()
      if n == peek:
        continue
      else:
        obj.push(n)
  return obj

print('Test 01')
st=Stack()
st.push(10)
st.push(10)
st.push(20)
st.push(20)
st.push(30)
st.push(10)
st.push(50)
print('Stack:')
print_stack(st)
print('------')
reversed_stack=conditional_reverse(st)
print('Conditional Reversed Stack:')
print_stack(reversed_stack) # This stack contains 50, 10, 30, 20, 10 in this order whereas top element should be 10
print('------')

```
