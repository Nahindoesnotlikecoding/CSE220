```python
import numpy as np
```

```python
class Node:
  def __init__(self, elem, weight=1, next=None):
    self.elem = elem
    self.weight = weight
    self.next = next
```

## Task 1

```python
class UndirectedGraph:
  def __init__(self, n):
    self.num_of_vertex = n+1
    self.adj = np.empty(self.num_of_vertex, dtype=Node)

  def addEdge(self, u, v):
    if self.adj[u] != None:
      temp = self.adj[u]
      while temp.next != None:
        temp = temp.next
      temp.next = Node(v)
    else:
      self.adj[u] = Node(v)

    if self.adj[v] != None:
      temp = self.adj[v]
      while temp.next != None:
        temp = temp.next
      temp.next = Node(u)
    else:
      self.adj[v] = Node(u)
  def show(self):
    for i in range(1, self.num_of_vertex):
      print(i, end=": ")
      temp = self.adj[i]
      while (temp != None):
        print(temp.elem, end=" --> ")
        temp = temp.next
      print("None")

def maximum_degree(graph):
  max = 0
  max_vertice = 0
  for i in range(graph.num_of_vertex):
    if graph.adj[i] != None:
      temp = graph.adj[i]
      num = 0
      while temp!=None:
        num += 1
        temp = temp.next

      if num >= max:
        max = num
        max_vertice = i

  return f"vertex with maximum degree is {max_vertice} and degree of that vertex is {max}"



G = UndirectedGraph(8)
G.addEdge(1, 2)
G.addEdge(1, 4)
G.addEdge(1, 5)
G.addEdge(1, 6)
G.addEdge(1, 7)
G.addEdge(2, 5)
G.addEdge(2, 6)
G.addEdge(2, 7)
G.addEdge(2, 3)
G.addEdge(3, 8)
G.addEdge(3, 5)
G.addEdge(3, 4)
G.addEdge(4, 5)
G.addEdge(5, 8)
G.addEdge(5, 6)
G.addEdge(6, 7)
G.addEdge(7, 8)
# G.show()
x = maximum_degree(G)
print(x)
```

```python
class UndirectedMatrix:
  def __init__(self, n):
    self.num_of_vertex = n
    self.matrix = np.zeros(shape=(n, n), dtype="int")

  def addEdge(self, u, v):
    self.matrix[u-1][v-1] = 1
    self.matrix[v-1][u-1] = 1

  def show(self):
    print(self.matrix)

def maximum_degree(graph):
  max = 0
  max_vertice = 0
  for i in range(graph.num_of_vertex):
    num = 0
    for j in range(graph.num_of_vertex):
      if graph.matrix[i][j] == 1:
        num += 1

    if num >= max:
      max = num
      max_vertice = i+1

  return f"vertex with maximum degree is {max_vertice} and degree of that vertex is {max}"

G = UndirectedMatrix(8)
G.addEdge(1, 2)
G.addEdge(1, 4)
G.addEdge(1, 5)
G.addEdge(1, 6)
G.addEdge(1, 7)
G.addEdge(2, 5)
G.addEdge(2, 6)
G.addEdge(2, 7)
G.addEdge(2, 3)
G.addEdge(3, 8)
G.addEdge(3, 5)
G.addEdge(3, 4)
G.addEdge(4, 5)
G.addEdge(5, 8)
G.addEdge(5, 6)
G.addEdge(6, 7)
G.addEdge(7, 8)

# G.show()
x = maximum_degree(G)
print(x)
```

```python
class DirectedWeightedGraph:
  def __init__(self, n):
    self.num_of_vertex = n+1
    self.adj = np.empty(self.num_of_vertex, dtype=Node)

  def addEdge(self, u, v, w):
    if self.adj[u] != None:
      temp = self.adj[u]
      while temp.next != None:
        temp = temp.next
      temp.next = Node(v, w)
    else:
      self.adj[u] = Node(v, w)

  def show(self):
    for i in range(1, self.num_of_vertex):
      print(i, end=": ")
      temp = self.adj[i]
      while (temp != None):
        print(temp.elem, end=" --> ")
        temp = temp.next
      print("None")

def convertToUndirected(graph):
  for i in range(1, graph.num_of_vertex):
    if graph.adj[i] != None:
      head = graph.adj[i]
      while head != None:
        temp = graph.adj[head.elem]
        exist = False
        while temp != None:
          if temp.elem == i:
            exist = True
            break
          temp = temp.next

        if exist != True:
          if graph.adj[head.elem] != None:
            temp2 = graph.adj[head.elem]
            while temp2.next != None:
              temp2 = temp2.next
            temp2.next = Node(i, head.weight)
          else:
            graph.adj[head.elem] = Node(i, head.weight)
        head = head.next

G = DirectedWeightedGraph(6)
G.addEdge(1, 2, 2)
G.addEdge(1, 3, 3)
G.addEdge(2, 6, 2)
G.addEdge(2, 5, 3)
G.addEdge(3, 6, 2)
G.addEdge(4, 5, 1)
G.addEdge(4, 2, 1)
G.addEdge(5, 6, 3)
G.addEdge(6, 1, 6)
G.show()

x = convertToUndirected(G)

G.show()
```
